```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
```

```python
# link : https://github.com/prashant-kikani/breast-cancer-detection/blob/master/breast-cancer-data.csv
df=pd.read_csv("breast-cancer-data.csv")
df.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0],test_size=0.2)
```

# Scale the data  to send in NN

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(X_train)
x_test=scaler.transform(X_test)
```

```python
y_train
```

# Now change label 

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
y_test
```

# Convert Numpy Array to pytorch tensor

```python
x_train_tensor=torch.from_numpy(x_train).to(torch.float32)
x_test_tensor=torch.from_numpy(x_test).to(torch.float32)
y_train_tensor=torch.from_numpy(y_train).to(torch.float32).unsqueeze(1)
y_test_tensor=torch.from_numpy(y_test).to(torch.float32).unsqueeze(1)
```

```python
x_train_tensor.shape
```

# Define Model

```python
class MyNN():
    def __init__(self,x):
        self.weights=torch.rand(x.shape[1],1,dtype=torch.float32,requires_grad=True)
        self.bias=torch.zeros(1,dtype=torch.float32,requires_grad=True)
    def forward(self,x):
        y=torch.matmul(x,self.weights)+self.bias
        y_predict=torch.sigmoid(y)
        return y_predict

    #Binanry cross Entrophy
    def binary_cross_entrophy(self,pre,tar):
        epsilon=1e-8   #to avoid log(0) we will create range btw 1e-8 to 1-1e-8
        pre=torch.clamp(pre,epsilon,1-epsilon)
        loss=-(tar*torch.log(pre)+(1-tar)*torch.log(1-pre)).mean()
        return loss


```

# Training Pipeline

```python
model = MyNN(x_train_tensor)
loss_fn = torch.nn.BCELoss()


lr = 0.01
epochs = 50

for epoch in range(epochs):

    # forward pass
    y_pre = model.forward(x_train_tensor)
    # print(y_pre.shape)
    # print(y_train_tensor.shape)

    # compute loss
    # loss = model.binary_cross_entrophy(y_pre, y_train_tensor)
    loss = loss_fn(y_pre, y_train_tensor)

    # backward pass
    loss.backward()

    # update weights
    with torch.no_grad():
        model.weights -= lr * model.weights.grad
        model.bias    -= lr * model.bias.grad

    # clear gradients
    model.weights.grad.zero_() 
    model.bias.grad.zero_()

    print(f"Epoch {epoch+1}, Loss = {loss.item():.4f}")
```

# Evaluation

```python
with torch.no_grad():
    y_pre=model.forward(x_test_tensor)
print((y_pre>0.5).float())
acc=(y_pre==y_test_tensor).float().mean()
print(f"Accuracy : {acc.item()}")
```
