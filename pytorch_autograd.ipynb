## Use simple NN

```python
# input
x=torch.tensor(6.7) #feature
y=torch.tensor(0.0) #true label

w=torch.tensor(1.0,requires_grad=True) #weight
b=torch.tensor(0.0,requires_grad=True) #bias

```

```python
#Binanry cross Entrophy
def binary_cross_entrophy(pre,tar):
    epsilon=1e-8   #to avoid log(0) we will create range btw 1e-8 to 1-1e-8
    pre=torch.clamp(pre,epsilon,1-epsilon)
    return -(tar*torch.log(pre)+(1-tar)*torch.log(1-pre))
```

```python
# forwad pass
z=w*x+b #weighted sum (linear)
y_pre=torch.sigmoid(z) # predicted prob
loss=binary_cross_entrophy(y_pre,y) # loss compute 
loss.backward()
print(w.grad) 
print(b.grad)
w.grad.zero_()
b.grad.zero_()
```
