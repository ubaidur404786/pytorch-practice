# Implace operstion 
when we want to change original tensor after operation 

```python
m=torch.rand(2,3)
n=torch.rand(2,3)
print(m)
print(n)
```

```python
m.add_(n)
```

```python
m #see the result update the original m
```

```python
# we cannot use in relu 
o=torch.rand(2,3)
print(o)
print(torch.relu(o))
print(o) # no effect  in original o tensor 

# now we will use implace method 
print(o.relu_())
print(o) # now changes apply in real o
```

# Copying a Tensor 

```python
p=torch.rand(2,3)
q=torch.rand(2,3)
print(p)
print(q)
```

```python
# we cannot use = sign for copy bcz then they will use same id reference if we apply chages on one tensor it will reflect in other to 
r=q
r[0][0]=0
print(r)
print(q)
print(id(r))
print(id(q))
```

```python
# So we will use clone for copy 
s=q.clone()
print(id(s))
print(id(q))
```

# Tensor Operation on GPU

```python
# check gpu available ?
print(torch.cuda.is_available())
```

```python
#create device
device=torch.device('cuda')
t=torch.rand((2,3),device=device)
# creating a new tensor on GPU
print(t)
```

```python
# moving existing tensor to GPU
v=s.to(device)
print(t)
print(v)

# now automatically each operation will use gpu like
print(t+v)
```

```python
# Compare bwt CPU vs GPU woth time 
import time

size=1000
m1_cpu=torch.rand(size,size)
m2_cpu=torch.rand(size,size)

# time measure
start_time=time.time()

result_cpu=torch.matmul(m1_cpu,m2_cpu)
cpu_time=time.time()-start_time

print("Time on cpu", cpu_time)

# device=torch.device('cuda')
m1_gpu=m1_cpu.to('cuda')
m2_gpu=m2_cpu.to('cuda')
# time measure
start_time=time.time()

result_gpu=torch.matmul(m1_gpu,m2_gpu)
torch.cuda.synchronize()
gpu_time=time.time()-start_time

print("Time on gpu", gpu_time)

print("speed up " ,cpu_time/gpu_time)
```

```python
b=torch.rand(2,3,4)
print(b.shape) # 3d shape index 0=2, index 1=3, index 2=4 example RGB
print(b)
```

```python
#permute change index of dimention
c=b.permute(2,0,1) #change index 
print(c.shape)
print(c)
```

```python
#unsqeeze , adding new dimetion 
d=torch.rand(226,226,3) #because in DNN model takes baches so we can add dim in 0 index like 
print(d.shape)
d.unsqueeze_(0) #we want to change in original d tensor so we used implace _ , this will be use full
print(d.shape)
d.unsqueeze_(1) #add in 1 index
print(d.shape)
d.unsqueeze_(2)  # add in 2nd index 
print(d.shape)

```

```python
#squeeze is opposite of unsqueeze to remove dim
print(d.squeeze_(0).shape)
#squeeze is opposite of unsqueeze to remove dim
print(d.squeeze_(0).shape)
#squeeze is opposite of unsqueeze to remove dim
print(d.squeeze_(0).shape) #now we have original one or  we can we remove channel from tensor like
print(d.squeeze_(2).shape) # this will not work because squeeze mean we can remove only that dimention whi has 1 in their diminton 
```

```python
print(d.unsqueeze_(2).shape) # add dim in 2 index

print(d.squeeze_(2).shape) # remove from 2 index 
```
