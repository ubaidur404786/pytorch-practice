```python
#Create model class 
import torch 
import torch.nn as nn
class Model(nn.Module):
    def __init__(self,num_features):
        super().__init__()
        self.linear=nn.Linear(num_features,1)
        self.sigmoid=nn.Sigmoid()

    def forward(self,features):
        out=self.linear(features)
        out=self.sigmoid(out)
        return out
```

```python
# create dataset 
features=torch.rand(10,5)
model=Model(features.shape[1])# intialize weights and bias 
pre=model(features)# its a magic method which will call forward method and will give predicted output
```

```python
print(pre)
```

```python
# check weights of linear layer
print(model.linear.weight)
print(model.linear.bias)
```
