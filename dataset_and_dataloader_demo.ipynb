```python
from sklearn.datasets import make_classification
import torch
```

```python
# Create synthetic classification dataset using sklearn
X,y=make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42
)
```

```python
X
```

```python
print(X.shape)
print(y)
print(type(X))
```

```python
#convert into a tensor
X=torch.tensor(X,dtype=torch.float32)
y=torch.tensor(y,dtype=torch.long)
print(type(X))
print(type(y))
```

```python
from torch.utils.data import Dataset,DataLoader
class CustomDataset(Dataset):
    def __init__(self,features,labels):
        self.features=features
        self.labels=labels
    def __len__(self):
        return self.features.shape[0]
    def __getitem__(self,index):
        return self.features[index],self.labels[index]

```

```python
dataset=CustomDataset(X,y)
```

```python
#lest get some methods values len , specific value 
print(len(dataset))
print(dataset[6])
#we can load the data and fetch the data from dataset 
```

```python
# now load the dataloader
# if __name__ == "__main__": #num_workers only run inside the main other wise will show error
dataloader=DataLoader(dataset,batch_size=20,shuffle=True,num_workers=0,pin_memory=True)
# pin_memory is true then gpu will on and fast the speed
```

```python
for batch_features,batch_labels in dataloader: # get randomly x , y batch iteratevely
    print(batch_features)
    print(batch_labels)
    print("-"*50)

    
```
